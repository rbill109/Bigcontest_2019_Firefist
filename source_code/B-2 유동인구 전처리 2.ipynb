# 노약자 유동인구 평균

```python
# 노원구
data = pd.read_csv("../data/유동인구/성연령별_노원구_유동인구.csv",encoding='euc-kr') 

# 노약자 계층 분리
data = data[["Date","Dong","M_0024","M_6064","M_6599","F_0024","F_6064","F_6599"]]

# 동별 일별 평균값 구하기
data["old_sum"] = data.iloc[:,2:].sum(axis=1)

# 날, 동, 일 구분
data_1 = data[["Date","Dong","old_sum"]]
# 1~3월 데이터 뽑아내기
data_2 = data_1.iloc[5225:]
data_2.reset_index(inplace=True)
del data_2["index"]
data_2
```

```python
# 동별 평균 데이터 뽑아내서 저장
groups = data_2.groupby(data_2.Dong)
pop_nflow = groups.mean()["old_sum"].reset_index()
pop_nflow
```

```python
# 종로구
# 데이터 불러오기
data = pd.read_csv("../data/유동인구/성연령별_종로구_유동인구.csv",encoding='euc-kr') 

# 노약자 계층 분리
data = data[["Date","Dong","M_0024","M_6064","M_6599","F_0024","F_6064","F_6599"]]

# 동별 일별 평균값 구하기
data["old_sum"] = data.iloc[:,2:].sum(axis=1)

# 날, 동, 일 구분
data_1 = data[["Date","Dong","old_sum"]]
# 1~3월 데이터 뽑아내기
data_2 = data_1.iloc[4675:]
data_2.reset_index(inplace=True)
del data_2["index"]
data_2
```

```python
# 동별 평균 데이터 뽑아내서 저장
groups = data_2.groupby(data_2.Dong)
pop_jflow = groups.mean()["old_sum"].reset_index()
pop_jflow
```

```python
nowon=pop_nflow
jonglo=pop_jflow
```

```python
nowon.set_index('Dong',inplace=True)
jonglo.set_index('Dong',inplace=True)
```

```python
# 노원구 노약자 유동인구 시각화
nowon.sort_values('old_sum').plot(kind='barh',grid=True,figsize=(10,6))
plt.show()
```

# 전체 유동인구 1~3월 동별 평균

```python
# 노원구
# 데이터 불러오기
data = pd.read_csv("../data/유동인구/성연령별_노원구_유동인구.csv",encoding='euc-kr') 

# 동별 일별 평균값 구하기
data["sum"] = data.iloc[:,2:].sum(axis=1)

# 날, 동, 일 구분
data_1 = data[["Date","Dong","sum"]]
data_2 = data_1.iloc[5225:]
data_2.reset_index(inplace=True)
del data_2["index"]
groups = data_2.groupby(data_2.Dong)
nowon_flowall= groups.mean()["sum"].reset_index()
nowon_flowall.head()
```

```python
# 종로구
# 데이터 불러오기
data = pd.read_csv("../data/유동인구/성연령별_종로구_유동인구.csv",encoding='euc-kr') 

# 동별 일별 평균값 구하기
data["sum"] = data.iloc[:,2:].sum(axis=1)

# 날, 동, 일 구분
data_1 = data[["Date","Dong","sum"]]
data_2 = data_1.iloc[5225:]
data_2.reset_index(inplace=True)
del data_2["index"]
groups = data_2.groupby(data_2.Dong)
jonglo_flowall = groups.mean()["sum"].reset_index()
jonglo_flowall.set_index('Dong',inplace=True)
nowon_flowall.set_index('Dong',inplace=True)
```

```python
n_df = pd.concat([nowon,nowon_flowall],axis=1)
j_df = pd.concat([jonglo,jonglo_flowall],axis=1)
```

```python
n_df['비율'] = n_df['old_sum']/n_df['sum']
```

```python
j_df['비율'] = j_df['old_sum']/j_df['sum']
```
