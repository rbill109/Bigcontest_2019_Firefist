```python
data = pd.read_excel('D:/DataScience/data/데이터_설명/04_Innovation 분야_환경기상데이터(케이웨더)_데이터정의서(행정동추가).xlsx',
                    sheet_name='경진대회용 후보 지점',header=0)
data.head()
```

```python
# 종로구

# 위치 뽑아내기
address=[]
for i in data.index:
    address.append(data['위치'][i])
# print(address)

# 종로구 관측소 주소를 geocode 함수 통해 위도, 경도 뽑아내서 각각 list에 넣어줌(추후 데이터프레임 작성 위해)
jonglo_address = [] # 주소
jonglo_lat = [] # 위도
jonglo_lng = [] # 경도

for addr in address:
    try:
        tmp = gmaps.geocode(addr,language='ko')
    #     print(tmp)
        jonglo_address.append(tmp[0].get('formatted_address'))
    #     print(station_address)
        tmp_loc = tmp[0].get('geometry')
        
        jonglo_lat.append(tmp_loc['location']['lat'])
        jonglo_lng.append(tmp_loc['location']['lng'])
    except:
        # '묘동'의 경우 에러가 나기 때문에 따로 위,경도 구하여 대입해줌
        jonglo_address.append('서울특별시 종로구 묘동 88-1')
        jonglo_lat.append(37.572522)
        jonglo_lng.append(126.991864)


# 위에서 만든 list를 활용하여 종로구 관측소 위경도 데이터 프레임 생성
jonglo = pd.DataFrame()
jonglo['주소']=jonglo_address
jonglo['위도']=jonglo_lat
jonglo['경도']=jonglo_lng
jonglo
```

```python
# 노원구

# 위치 뽑아내기
address2=[]
for i in range(22):
    address2.append(data['위치.1'][i])
# print(address2)


# 노원구 관측소 주소를 geocode 함수 통해 위도, 경도 뽑아내서 각각 list에 넣어줌(추후 데이터프레임 작성 위해)
nowon_address = [] # 주소
nowon_lat = [] # 위도
nowon_lng = [] # 경도
for addr in address2:
    try:
        tmp = gmaps.geocode(addr,language='ko')
    #     print(tmp)
        nowon_address.append(tmp[0].get('formatted_address'))
    #     print(station_address)
        tmp_loc = tmp[0].get('geometry')
        nowon_lat.append(tmp_loc['location']['lat'])
        nowon_lng.append(tmp_loc['location']['lng'])
    except:
        # 종로구와 달리 노원구는 에러값이 없으므로 패스함
        pass

    
# 위에서 만든 list를 활용하여 종로구 관측소 위경도 데이터 프레임 생성
nowon = pd.DataFrame()
nowon['주소']=nowon_address
nowon['위도']=nowon_lat
nowon['경도']=nowon_lng
nowon
```

# 3. 관측소 없는 동 뽑아내어 가장 가까운 관측소 선정하기

```python
# 관측소 데이터를 활용하여 행정동 위치 뽑아내기
jr_dong_list = list(data['행정동'].unique())
nw_dong_list = list(data['행정동.1'].unique())
# print(jr_dong_list,nw_dong_list)

# 노원 행정동 마지막 원소 제거(nan값)
del nw_dong_list[-1]

# 행정동명 통합 (,와 공백 제거)
for i in range(len(jr_dong_list)) :
    jr_dong_list[i] = jr_dong_list[i].replace(" ","").replace(",","")
for i in range(len(nw_dong_list)) :
    nw_dong_list[i] = nw_dong_list[i].replace(" ","").replace(",","")

# 전체 동 데이터 리스트로 가져오기
j_dong = ['청운효자동', '사직동', '삼청동', '부암동', '평창동', '무악동', '교남동', '가회동',
       '종로1234가동', '종로56가동', '이화동', '혜화동', '창신1동', '창신2동', '창신3동',
       '숭인1동', '숭인2동']
n_dong = ['월계1동','월계2동','월계3동',
        '공릉1동','공릉2동',
        '하계1동','하계2동',
        '중계본동','중계1동','중계4동','중계23동',
        '상계1동','상계2동','상계34동','상계5동','상계67동','상계8동','상계9동','상계10동']
```

```python
# 가진 행정동과 기준 행정동 각각 데이터프레임으로 만들어줌
# concat함수로 없는 내용 nan값 주기 위해 동이름을 각각 index로 설정
jr_dong_origin = pd.DataFrame({"dong":j_dong,
                             "value":["True" for i in range(len(j_dong))]}).set_index("dong")
jr_dong_have = pd.DataFrame({"dong":jr_dong_list,
                            "value_check":["True" for i in range(len(jr_dong_list))]}).set_index("dong")
nw_dong_origin = pd.DataFrame({"dong":n_dong,
                             "value":["True" for i in range(len(n_dong))]}).set_index("dong")
nw_dong_have = pd.DataFrame({"dong":nw_dong_list,
                            "value_check":["True" for i in range(len(nw_dong_list))]}).set_index("dong")

# 종로구 비교 데이터프레임
jr_compare = pd.concat([jr_dong_origin,jr_dong_have],axis=1)
# 노원구 비교 데이터프레임
nw_compare = pd.concat([nw_dong_origin,nw_dong_have],axis=1)
```

```python
# 종로구
# 필요한데 없는 동
jr_missing = list(jr_compare.loc[jr_compare.isnull()["value_check"],:].index)
# > 무악동, 삼청동, 숭인1동, 창신2동

# 필요없는데(종로구 아닌데) 포함된 동
jr_compare.loc[jr_compare.isnull()["value"],:].index
# > 명동, 천연동


# 노원구
# 필요한데 없는 동
nw_missing = list(nw_compare.loc[nw_compare.isnull()["value_check"],:].index)
# > 상계8동, 상계9동, 월계2동, 중계1동, 중계4동, 하계2동

# 필요없는데(노원구 아닌데) 포함된 동
nw_compare.loc[nw_compare.isnull()["value"],:].index
# > 없음
```

```python
# jr_missing과 nw_missing에 포함된 동들은 관측소가 없는 동임. 따라서, 가장 가까운 관측소를 선정할 것
# 각 동의 위경도 값은 중앙 위치를 임의로 선정하여 부여함
jr_missing_df = pd.DataFrame({"관측소":jr_missing,
                             "위도":[37.577336,37.584873,37.577594,37.576069],
                             "경도":[126.958627,126.981957,127.016920,127.010306]})
nw_missing_df = pd.DataFrame({"관측소":nw_missing,
                             "위도":[37.666026,37.666782,37.633551,37.650191,37.658571,37.634768],
                             "경도":[127.053941,127.062885,127.051293,127.073643,127.085022,37.634768]})

# 위경도 값을 입력하면 주어진 관측소 정보 중 가장 가까운 정보를 반환하는 함수 생성
def find_closest(lat,lng,jr_nw) :
    from haversine import haversine
    dis = []
    reg = []
    for i in range(0,len(jr_nw)): # 종로,노원 위치함수 합치기
        lyon = (lat, lng) 
        paris = (jr_nw.iloc[i,1], jr_nw.iloc[i,2])
        a = haversine(lyon, paris)
        b = jr_nw.iloc[i,0]
        dis.append(a)
        reg.append(b)
    return pd.DataFrame({"지역":reg, "거리":dis}).sort_values("거리").head(1)

# 한번에 확인하기 위해 종로와 노원 데이터 합쳐놓기
jr_nw_df = pd.concat([jonglo,nowon],ignore_index=True)

# 함수 test
find_closest(jr_missing_df.iloc[0,:][1],jr_missing_df.iloc[0,:][2],jr_nw_df)
```

```python
# 가장 가까운 정류소 주소 구하기

for i in range(len(jr_missing_df["관측소"])) :
    result = find_closest(jr_missing_df.iloc[i,:][1],jr_missing_df.iloc[i,:][2],jr_nw_df)
    print(str(jr_missing_df.iloc[i,:][0]))
    print("에서 가장 가까운 정류소 주소는")
    print(str(result["지역"]),"\n")
    
for i in range(len(nw_missing_df["관측소"])) :
    result = find_closest(nw_missing_df.iloc[i,:][1],nw_missing_df.iloc[i,:][2],jr_nw_df)
    print(str(nw_missing_df.iloc[i,:][0]))
    print("에서 가장 가까운 정류소 주소는")
    print(str(result["지역"]),"\n")
```
