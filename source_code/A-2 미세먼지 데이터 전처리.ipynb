# 미세먼지 데이터 가공 함수 
### 시간별 데이터 ▶ 결측치 및 이상치 제거 ▶ 일별 평균 데이터로 가공

```python
import pandas as pd

def dust_mean(raw) :
    date_list = []
    hour_list = []
    min_list = []
    for i in raw.tm :
        date_list.append(str(i)[:8])
    for i in raw.tm :
        hour_list.append(str(i)[8:10])
    for i in raw.tm :
        min_list.append(str(i)[10:])

    # 리스트 세개를 합쳐 새로운 dataframe 생성
    split_df = pd.DataFrame({"date":date_list,
                            "hour":hour_list,
                            "min":min_list})

    # pm10 열만 가져와서 병합
    raw = raw.loc[:,'pm10'] 
    result = pd.concat([split_df,raw], axis=1)
    
    # 결측치(-999, -9999) 제거
    result = result.loc[(result.pm10!=-999)&(result.pm10!=-9999),:]
    
    # 이상치 제거
    df = result['pm10'].describe()
    top_out = df[-2]+(df[-2]-df[-3])*1.5
    down_out = df[4]-(df[-2]-df[-3])*1.5
    result = result.loc[(result.pm10>=down_out)&(result.pm10<=top_out),:]

    grouped = result.groupby(result.date)
    result_date_mean = grouped.mean()
    
    return result_date_mean
```

#### 특정 폴더 안에 있는 raw 데이터를 읽어들여 데이터 프레임 생성

```python
# 파일경로 대입하면 데이터 프레임 생성하는 함수 생성
def read_file(file_path) :
    raw = pd.read_csv(file_path)
    return dust_mean(raw)
```

```python
# pm10이 80 이상인 경우 1로 범주화
def bad_day(df):
    df['bad'] = 0
    n = 0
    for i in df['pm10']:
        if i >= 80:
            df.iloc[n,1] = 1
        else:
            df.iloc[n,1] = 0
        n += 1
    df = df.reset_index()
    return df
# 모든 데이터프레임 범주화
for i in name_n:
    exec("bad_day(%s)" %i)
```

```python
# 월별 bad 빈도 수
def month_count(day):
    n = 0
    for i in day.index:
        day.iloc[n,0] = i[:6]
        n += 1
    df = pd.DataFrame(day['bad'].groupby(day['pm10']).sum().values,
                     index = day['bad'].groupby(day['pm10']).sum().index)
    return df
```

```python
# 분기별 bad 빈도 수
def quater_count(day):
    day.reset_index(inplace=True)
    n = 0
    for i in day['date']:
        if (i[4:6]=='01')|(i[5:6]=='02')|(i[5:6]=='03'):
            day.iloc[n,0] = 1
        elif (i[4:6]=='04')|(i[5:6]=='05')|(i[5:6]=='06'):
            day.iloc[n,0] = 2
        elif (i[4:6]=='07')|(i[5:6]=='08')|(i[5:6]=='09'):
            day.iloc[n,0] = 3
        else:
            day.iloc[n,0] = 4
        n += 1
    
    return day['bad'].groupby(day['date']).sum()
```
